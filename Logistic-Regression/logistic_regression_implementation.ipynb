```python
import seaborn as sns 
import pandas as pd 
import numpy as np
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
print(df['species'].unique())
print(df['species'].isnull().sum())
```

```python
df['species'].unique()
```

```python
# Logistic Regression works best for Binary Classification, So Let us take two output feature
df = df[df['species'] != 'setosa']
```

```python
df.head()
```

```python
#Converting the output versicolor and verginica into 0 and 1 
df['species']=df['species'].map({'versicolor':0,'virginica':1})
```

```python
df.head()
```

```python
# splitting dataset into independent and dependent features
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection  import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'penalty':['l1','l2','l3'], 'C':[1,2,3,4,5,8,9,10,20,30,40,50,60], 'max_iter':[100,200,300]}
```

```python
classifier_reg = GridSearchCV(classifier,param_grid=parameters,scoring='accuracy',cv=5)
```

```python
classifier_reg.fit(X_train,y_train)
```

```python
print(classifier_reg.best_params_)
```

```python
print(classifier_reg.best_score_)
```

```python
y_pred = classifier_reg.predict(X_test)
```

```python
# accuracy score
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred,y_test)
```

```python
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

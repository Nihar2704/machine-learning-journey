```python
from io import StringIO, BytesIO
```

```python
#creating a csv file with standard structure
data = ('col1,col2,col3\n'
       '2,3,1\n'
       '6,5,7\n'
       '9,0,8')
```

```python
type(data)
```

```python
pd.read_csv(StringIO(data)) #convert the created csv data which is in string format into a csv format
```

```python
my_df=pd.read_csv(StringIO(data), usecols=['col1','col3']) #usecols is used to retieve values from particular columns
```

```python
my_df.to_csv('Sample.csv') #converts the created csv data into a file
```

```python
my_df1 = pd.read_csv(StringIO(data), dtype=float) # we can also define the data type of dataframe
```

```python
my_df1
```

```python
my_df1['col1'][2] # accessing element 2 from column1
```

```python
my_df2 = pd.read_csv(StringIO(data), dtype={'col1':float, 'col2':int}) #we can also set different datatypes for columns using dtype and by defining the datatypes of particular columns
```

```python
my_df2
```

```python
my_df2.dtypes #it is used to check the datatypes of all the elements in dataframe
```

```python
new_data = ('index,c1,c2,c3\n'
           '2,John,Wwe,17\n'
           '3,Batista,Wwe,14')
```

```python
pd.read_csv(StringIO(new_data))
```

```python
pd.read_csv(StringIO(new_data), index_col=0) #index_col is used for indexing suppose in above example index is by default shown as 0 1 and i want to take value of index column as index value then it is used 
```

```python
new_data1 = ('a,b,c\n'
           '4,apple,bat,\n'
           '3,orange,cow,')
```

```python
pd.read_csv(StringIO(new_data1))
```

```python
pd.read_csv(StringIO(new_data1), index_col=False) #if index_col is set to false then it will take the elements as per the columns 
```

```python
#combining usecols and index_col
new_data1 = ('a,b,c\n'
           '4,apple,bat,\n'
           '3,orange,cow,')
```

```python
pd.read_csv(StringIO(new_data1), usecols=['b','c'], index_col=False) # using usecols and index_cols together
```

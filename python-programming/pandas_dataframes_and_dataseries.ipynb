Dataframe - It is a combination of multiple rows and columns 

DataSeries - It is one row or column

```python
# Ways to Access the dataframe elements
# 1) loc dataframe.loc[]
# 2) iloc dataframe.iloc[start:end,start:end] here you will have to give value of rows and columns
```

```python
df.loc['R2']  #it will show the elements if R2
```

```python
df.iloc[0:3,:]
```

```python
df.iloc[0:3,0:1]
```

```python
type(df.loc['R1'])  # loc will give elements that is data series
```

```python
type(df.iloc[0:2,:]) # iloc will give elements in dataframe
```

```python
df.iloc[0:4,0]
```

How to check null condition in a DataFrame

```python
df.isnull().sum() # it will show the number of null values in data frame
```

```python
df['C1'].unique() #it will give the unique values from column 1
```

```python
df['C1'].value_counts() #it will count the frequency of number of elements in a column
```
